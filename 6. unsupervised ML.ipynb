# **1Ô∏è‚É£ Clustering Algorithms**
Clustering is a technique to group similar data points based on their features.  

## **üîπ K-Means Clustering**
‚úÖ **How it works:**  
- Select **K clusters** randomly.  
- Assign each point to the nearest cluster.  
- Compute the new cluster centroids.  
- Repeat until clusters remain stable.  

### **üìå Code Example: K-Means on Synthetic Data**


```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic dataset
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label="Centroids")
plt.legend()
plt.title("K-Means Clustering")
plt.show()
```


**Key Observations:**  
- K-Means works well when clusters are clearly separable.  
- It may fail if clusters have **different sizes or densities**.  

---

## **üîπ DBSCAN (Density-Based Clustering)**
‚úÖ **How it works:**  
- Identifies **high-density** regions.  
- Points in **low-density** areas are marked as **outliers**.  
- No need to specify the number of clusters (unlike K-Means).  

### **üìå Code Example: DBSCAN on Non-Spherical Clusters**


```python

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generate dataset with non-spherical clusters
X, _ = make_moons(n_samples=300, noise=0.1, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma', edgecolors='k')
plt.title("DBSCAN Clustering")
plt.show()
```


**Key Observations:**  
- Works well for **non-spherical** clusters.  
- Automatically finds the number of clusters.  
- Struggles with **varying densities**.  

---

## **üîπ Hierarchical Clustering**
‚úÖ **How it works:**  
- Builds a **tree-like structure** of clusters (dendrogram).  
- No need to predefine the number of clusters.  
- Two types:  
  - **Agglomerative:** Bottom-up merging of clusters.  
  - **Divisive:** Top-down splitting of clusters.  

### **üìå Code Example: Agglomerative Clustering**


```python

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic dataset
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=0.6, random_state=42)

# Compute the linkage matrix
Z = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
y_agg = agg_clustering.fit_predict(X)

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=y_agg, cmap='coolwarm', edgecolors='k')
plt.title("Agglomerative Clustering")
plt.show()
```


**Key Observations:**  
- Provides **better interpretability** (dendrogram).  
- Computationally expensive for **large datasets**.

# **2Ô∏è‚É£ Dimensionality Reduction**
High-dimensional data can be **difficult to visualize and process**. Dimensionality reduction methods help extract meaningful features.  

## **üîπ Principal Component Analysis (PCA)**
‚úÖ **How it works:**  
- Identifies **directions of maximum variance** in data.  
- Projects data onto these directions to reduce dimensions.  

### **üìå Code Example: PCA for Dimensionality Reduction**


```python

from sklearn.decomposition import PCA
from sklearn.datasets import load_digits

# Load high-dimensional dataset (handwritten digits)
digits = load_digits()
X, y = digits.data, digits.target

# Apply PCA (reduce to 2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot reduced dataset
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Spectral', edgecolors='k')
plt.colorbar(label="Digit Label")
plt.title("PCA on Handwritten Digits")
plt.show()
```


**Key Observations:**  
- Reduces high-dimensional data to **fewer features**.  
- Useful for **visualization and speeding up models**.

# **3Ô∏è‚É£ Anomaly Detection**
Detects **unusual patterns** that do not conform to expected behavior.  

## **üîπ Isolation Forest**
‚úÖ **How it works:**  
- Randomly selects **features** and **splits data** into small regions.  
- Anomalies are **isolated quickly** (fewer splits needed).  

### **üìå Code Example: Detecting Anomalies with Isolation Forest**


```python

from sklearn.ensemble import IsolationForest

# Generate synthetic dataset
X = np.random.randn(300, 2)
X[:10] = X[:10] + 5  # Add anomalies

# Train Isolation Forest
iso_forest = IsolationForest(contamination=0.05)
y_pred = iso_forest.fit_predict(X)

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolors='k')
plt.title("Anomaly Detection with Isolation Forest")
plt.show()
```


**Key Observations:**  
- Useful for fraud detection, cybersecurity, and industrial monitoring.

# **1Ô∏è‚É£ Clustering (Grouping Similar Data)**
### **üìå Definition**
Clustering algorithms group similar data points together **without predefined labels**.

## **üîπ K-Means Clustering**
‚úÖ **How it works:**  
1. Select **K random cluster centroids**.  
2. Assign each data point to the nearest centroid.  
3. Recalculate centroids as the mean of the assigned points.  
4. Repeat until centroids don‚Äôt change significantly.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Easy to implement | Requires choosing K manually |
| Works well for spherical clusters | Struggles with non-spherical data |

---

## **üîπ DBSCAN (Density-Based Clustering)**
‚úÖ **How it works:**  
- Identifies **high-density regions** and labels points in sparse areas as **outliers**.  
- Unlike K-Means, it **does not require specifying K**.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Works for **non-spherical clusters** | Struggles with varying densities |
| Handles noise and outliers | Sensitive to hyperparameters |

---

## **üîπ Hierarchical Clustering**
‚úÖ **How it works:**  
- Builds a **tree of clusters (dendrogram)**.  
- Clusters are either **merged (agglomerative)** or **split (divisive)**.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| No need to predefine K | Computationally expensive for large data |
| Dendrogram provides better interpretability | Cannot handle noise well |

---

## **üîπ Gaussian Mixture Model (GMM)**
‚úÖ **How it works:**  
- Uses **probabilistic models** to cluster data.  
- Each cluster is represented by a **Gaussian distribution**.  
- Unlike K-Means, it assigns **soft probabilities** to each data point.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Works well for overlapping clusters | Requires tuning number of Gaussians |
| Soft clustering (data points can belong to multiple clusters) | Computationally expensive |

---



# **2Ô∏è‚É£ Dimensionality Reduction (Feature Extraction)**
### **üìå Definition**
Dimensionality reduction techniques **reduce high-dimensional data** into **fewer features** while retaining meaningful information.

## **üîπ Principal Component Analysis (PCA)**
‚úÖ **How it works:**  
- Identifies **directions of maximum variance** in the data.  
- Projects data onto these new axes (principal components).  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Reduces computation time | Loses interpretability |
| Useful for visualization | Works best for linear data |

---

## **üîπ t-SNE (t-Distributed Stochastic Neighbor Embedding)**
‚úÖ **How it works:**  
- Maps high-dimensional data into **2D/3D** using probabilities.  
- Maintains **local structure** in the data.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Best for visualization | Computationally expensive |
| Preserves local structure | Non-deterministic (results change per run) |

---

## **üîπ UMAP (Uniform Manifold Approximation and Projection)**
‚úÖ **How it works:**  
- Similar to t-SNE but **faster and better preserves global structure**.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Faster than t-SNE | Requires fine-tuning |
| Preserves more structure than t-SNE | May be less interpretable |

---



# **3Ô∏è‚É£ Anomaly Detection (Finding Outliers)**
### **üìå Definition**
Anomaly detection finds **rare, abnormal patterns** in data.  

## **üîπ Isolation Forest**
‚úÖ **How it works:**  
- Builds random trees where **anomalies are isolated quickly**.  
- Points that require **fewer splits** are classified as **anomalies**.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Works well for high-dimensional data | Sensitive to contamination parameter |
| Fast and scalable | May fail if anomalies are not well-separated |

---

## **üîπ One-Class SVM**
‚úÖ **How it works:**  
- Learns a **hyperplane** that separates normal data from anomalies.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Works well for **small datasets** | Not scalable for big data |
| Effective when anomalies are well-defined | Requires tuning hyperparameters |

---



# **4Ô∏è‚É£ Association Rule Learning (Finding Patterns in Data)**
### **üìå Definition**
These algorithms discover **relationships between items** in a dataset.  

## **üîπ Apriori Algorithm**
‚úÖ **How it works:**  
- Finds **frequent itemsets** based on **support, confidence, and lift**.  
- Used in **market basket analysis** (e.g., ‚ÄúPeople who buy milk also buy bread‚Äù).  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Works well for market basket analysis | Computationally expensive for large datasets |
| Simple and interpretable | Requires fine-tuning support threshold |

---

## **üîπ FP-Growth Algorithm**
‚úÖ **How it works:**  
- Uses a **tree-based approach** to find frequent patterns **faster than Apriori**.  

### **üìå Pros & Cons**
| ‚úÖ Advantages | ‚ùå Disadvantages |
|--------------|-----------------|
| Faster than Apriori | Needs preprocessing |
| Scales well for large data | Complex implementation |

Let's code this in Python. üöÄ


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
```


### **üìå What This Code Does**
1. **Loads the Mall Customer Dataset** and selects relevant features.  



```python
# Load Dataset
df = pd.read_csv("Mall_Customers.csv")
df = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
df.head()
```


2. **Standardizes the Data** for better clustering performance.  


```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
```

3. **Applies Clustering Algorithms**  
   - K-Means  
   - DBSCAN  
   - Hierarchical  
   - Gaussian Mixture Model (GMM)  


```python
# K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(data_scaled)

# DBSCAN Clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(data_scaled)

# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
df['Hierarchical_Cluster'] = hierarchical.fit_predict(data_scaled)

# Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=4, random_state=42)
df['GMM_Cluster'] = gmm.fit_predict(data_scaled)
```

4. **Uses PCA & t-SNE for Visualization**  


```python
# PCA for Visualization
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
df['PCA1'] = data_pca[:, 0]
df['PCA2'] = data_pca[:, 1]

# t-SNE for Visualization
tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(data_scaled)
df['TSNE1'] = data_tsne[:, 0]
df['TSNE2'] = data_tsne[:, 1]
```

5. **Applies Anomaly Detection** (Isolation Forest & One-Class SVM).  


```python
# Anomaly Detection
isolation_forest = IsolationForest(contamination=0.05, random_state=42)
df['IsolationForest_Anomaly'] = isolation_forest.fit_predict(data_scaled)

one_class_svm = OneClassSVM(nu=0.05)
df['OneClassSVM_Anomaly'] = one_class_svm.fit_predict(data_scaled)

```

6. **Plots Results** using PCA components.  


```python
# Plot Clustering Results
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='PCA1', y='PCA2', hue='KMeans_Cluster', palette='tab10', data=df)
plt.title('K-Means Clustering')

plt.subplot(1, 2, 2)
sns.scatterplot(x='PCA1', y='PCA2', hue='DBSCAN_Cluster', palette='tab10', data=df)
plt.title('DBSCAN Clustering')
plt.show()

# Print Final DataFrame with Cluster Labels
df.head()
```

### **üìå Comparing Unsupervised Learning Models**  
Unlike supervised learning (which uses accuracy, MSE, or other loss functions), **unsupervised learning models** are evaluated differently. Since we don‚Äôt have ground truth labels, we use **internal** and **external evaluation metrics**.

---

## **‚úÖ Metrics to Compare Clustering Models**
| Metric | Description | Best For |
|--------|------------|----------|
| **Inertia (WCSS - Within-Cluster Sum of Squares)** | Measures how tightly data points are clustered. Lower is better. | **K-Means** |
| **Silhouette Score** | Measures how well-separated clusters are. Higher is better (range: -1 to 1). | **All clustering models** |
| **Davies-Bouldin Index (DBI)** | Measures cluster separation and compactness. Lower is better. | **All clustering models** |
| **Calinski-Harabasz Index (CHI)** | Measures ratio of between-cluster dispersion to within-cluster dispersion. Higher is better. | **All clustering models** |

---

## **üìå Let‚Äôs Compare Our Models Using Python Code** üöÄ



```python

from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Function to evaluate clustering models
def evaluate_clustering(data, labels, model_name):
    silhouette = silhouette_score(data, labels)
    db_index = davies_bouldin_score(data, labels)
    ch_index = calinski_harabasz_score(data, labels)
    
    print(f"\nüìå {model_name} Clustering Performance:")
    print(f"‚úÖ Silhouette Score: {silhouette:.3f} (Higher is better)")
    print(f"‚úÖ Davies-Bouldin Index: {db_index:.3f} (Lower is better)")
    print(f"‚úÖ Calinski-Harabasz Index: {ch_index:.3f} (Higher is better)")

# Evaluate K-Means
evaluate_clustering(data_scaled, df['KMeans_Cluster'], "K-Means")

# Evaluate DBSCAN (Only if clusters are found)
if len(set(df['DBSCAN_Cluster'])) > 1:
    evaluate_clustering(data_scaled, df['DBSCAN_Cluster'], "DBSCAN")
else:
    print("\n‚ö†Ô∏è DBSCAN found only one cluster, evaluation skipped.")

# Evaluate Hierarchical Clustering
evaluate_clustering(data_scaled, df['Hierarchical_Cluster'], "Hierarchical")

# Evaluate Gaussian Mixture Model (GMM)
evaluate_clustering(data_scaled, df['GMM_Cluster'], "GMM")
```



---

### **üìå How to Interpret the Results?**
- **Higher Silhouette Score (closer to 1)** ‚Üí Better cluster separation.  
- **Lower Davies-Bouldin Index** ‚Üí Better-defined clusters.  
- **Higher Calinski-Harabasz Index** ‚Üí Higher cluster dispersion (better clustering).

To **compare clustering models**, we'll use **Principal Component Analysis (PCA)** and **t-SNE** to reduce dimensions and plot the clusters.

## **‚úÖ Step 1: Import Required Libraries**  


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Set plot style
sns.set(style="whitegrid")
```



---

## **‚úÖ Step 2: Define Visualization Function**


```python

def plot_clusters(data, labels, title):
    """Reduces data using PCA and plots clusters"""
    pca = PCA(n_components=2)  # Reduce to 2D for visualization
    reduced_data = pca.fit_transform(data)
    
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap="viridis", alpha=0.7)
    plt.colorbar(scatter)
    plt.title(f"üìå {title} Clustering Results")
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.show()
```



---

## **‚úÖ Step 3: Plot Clusters for Each Model**


```python

# Visualize K-Means Clustering
plot_clusters(data_scaled, df['KMeans_Cluster'], "K-Means")

# Visualize DBSCAN Clustering (Skip if only 1 cluster)
if len(set(df['DBSCAN_Cluster'])) > 1:
    plot_clusters(data_scaled, df['DBSCAN_Cluster'], "DBSCAN")
else:
    print("‚ö†Ô∏è DBSCAN found only one cluster, visualization skipped.")

# Visualize Hierarchical Clustering
plot_clusters(data_scaled, df['Hierarchical_Cluster'], "Hierarchical")

# Visualize Gaussian Mixture Model (GMM)
plot_clusters(data_scaled, df['GMM_Cluster'], "Gaussian Mixture Model (GMM)")
```



---

## **‚úÖ Step 4: Use t-SNE for Non-Linear Visualization**
PCA is **linear**, but **t-SNE** captures non-linear relationships better.



```python

def plot_tsne(data, labels, title):
    tsne = TSNE(n_components=2, perplexity=30, random_state=42)
    tsne_results = tsne.fit_transform(data)
    
    plt.figure(figsize=(8,6))
    scatter = plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=labels, cmap="coolwarm", alpha=0.7)
    plt.colorbar(scatter)
    plt.title(f"üìå {title} Clustering Results (t-SNE)")
    plt.xlabel("t-SNE Component 1")
    plt.ylabel("t-SNE Component 2")
    plt.show()

# Visualize using t-SNE
plot_tsne(data_scaled, df['KMeans_Cluster'], "K-Means")
plot_tsne(data_scaled, df['Hierarchical_Cluster'], "Hierarchical")
plot_tsne(data_scaled, df['GMM_Cluster'], "Gaussian Mixture Model (GMM)")
```
